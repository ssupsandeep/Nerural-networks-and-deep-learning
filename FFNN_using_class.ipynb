```python
import numpy as np

class FeedForwardNN:
    def __init__(self):
        self.a = np.array([0.3, 0.5, 0.4])
        self.W1 = np.array([[0.2, 0.3],
                            [0.1, 0.4],
                            [0.5, 0.7]])
        self.b1 = np.array([0.7, 0.2])

        self.W2 = np.array([[0.3],
                            [0.5]])
        self.b2 = np.array([0.05])

        self.lr = 0.1

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def forward(self):
        self.z1 = np.dot(self.a, self.W1) + self.b1
        self.a1 = self.sigmoid(self.z1)

        self.z2 = np.dot(self.a1, self.W2) + self.b2
        self.a2 = self.sigmoid(self.z2)

        return self.a2

    def compute_loss(self, target):
        self.loss = (target - self.a2) ** 2
        return self.loss

    def backward(self, target):
        dloss_da2 = -2 * (target - self.a2)
        da2_dz2 = self.a2 * (1 - self.a2)

        dz2_dW2 = self.a1.reshape(2, 1)

        grad_W2 = dloss_da2 * da2_dz2 * dz2_dW2

        self.W2 = self.W2 - self.lr * grad_W2

    def train_step(self, target):
        output = self.forward()
        loss = self.compute_loss(target)
        self.backward(target)

        return output, loss


nn = FeedForwardNN()

target = 1

epochs = 1000

for epoch in range(epochs):
    output, loss = nn.train_step(target)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Output={output[0]:.4f}, Loss={loss[0]:.6f}")
```
