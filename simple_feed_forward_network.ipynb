```python
import numpy as np
a=np.array([0.3,0.5,0.4])
b=np.array([[0.2,0.3],[0.1,0.4],[0.5,0.7]])
c=np.array([0.7,0.2])

x=np.dot(a,b)+c
print("Output:",x)
def sigmoid(x):
  return 1/(1+np.exp(-x))

y=sigmoid(x)
print("Activation output:",y)




w2=np.array([[0.3],[0.5]])
a1=np.array([0.05])

b2=np.dot(y, w2) + a1
a2=sigmoid(x)
print("Final output:",a2)
z=1
loss=(z-a2)**2
print("Loss:",loss)
lr=0.1
dloss_da2=-2 * (z-a2)
da2_dz2=a2 * (1-a2)
dz2_dw2=y

grad_w2=dloss_da2 * da2_dz2 * dz2_dw2
w2=w2-lr*grad_w2.reshape(2,1)
print("Updated W2:",w2)
```
