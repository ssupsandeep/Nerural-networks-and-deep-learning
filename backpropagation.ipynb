```python
import numpy as np
# -------- Input layer --------
# Input vector (different values)
a = np.array([0.6, 0.2, 0.1])
# Weights from input layer to hidden layer
b = np.array([[0.4, 0.1],
            [0.6, 0.3],
            [0.2, 0.5]])
# Bias for hidden layer
c = np.array([0.3, 0.6])
# -------- Hidden layer --------
# Linear combination (weighted sum + bias)
x = np.dot(a, b) + c
print ("Hidden layer linear output:", x)
# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
# Activation output of hidden layer
y = sigmoid(x)
print ("Hidden layer activation output:", y)
# -------- Output layer --------
# Weights from hidden layer to output layer
w2 = np.array([[0.4],
            [0.2]])
# Bias for output layer
a1 = np.array([0.1])
# Linear combination at output layer
b2 = np.dot(y, w2) + a1
# Final output using sigmoid activation
a2 = sigmoid(b2)
print ("Final output:", a2)
# -------- Loss calculation --------
# Target output (different target)
z = 0
# Mean Squared Error (MSE) loss
loss = (z - a2) ** 2
print ("Loss:", loss)
# -------- Backpropagation --------
# Learning rate
lr = 0.1
# Derivative of loss w.r.t output
dloss_da2 = -2 * (z - a2)
# Derivative of sigmoid
da2_dz2 = a2 * (1 - a2)
# Derivative of z2 w.r.t weights w2
dz2_dw2 = y
# Gradient of loss w.r.t w2
grad_w2 = dloss_da2 * da2_dz2 * dz2_dw2
# Update weights using gradient descent
w2 = w2 - lr * grad_w2.reshape(2, 1)
print ("Updated W2:", w2)
```

```python
import numpy as np
import tensorflow as tf
# Input data (different values)
X = np.array([[0.6, 0.2, 0.1]], dtype=np.float32)
# Target output (different target)
y = np.array([[0.0]], dtype=np.float32)
# Define a simple neural network model
model = tf.keras.Sequential([
tf.keras.layers.Dense(
3, # Different number of hidden neurons
activation='sigmoid',
input_shape=(3,)),
tf.keras.layers.Dense(
1,
activation='sigmoid')])
# Compile the model with a different learning rate
model.compile(
optimizer=tf.keras.optimizers.SGD(learning_rate=0.05),
loss='mse')
# Train the model for fewer epochs
history = model.fit(X,y,epochs=800,verbose=0)
# Predict output
output = model.predict(X)
print ("Final Output:", output)
```
